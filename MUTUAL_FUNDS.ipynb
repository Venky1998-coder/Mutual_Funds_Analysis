```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
MF=pd.read_csv(r'C:\Users\venka\OneDrive\Documents\MUTUAL_FUNDS.csv')
```

```python
MF.info()
```

```python
MF.head()
```

```python
plt.hist(MF['expense_ratio'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Expense Ratio')
plt.ylabel('Frequency')
plt.title('Histogram of Expense Ratio')
plt.show()
```

```python
sns.boxplot(y=MF['fund_size_cr'])
plt.ylabel('Fund Size (in Crores)')
plt.title('Boxplot of Fund Size')
plt.show()
```

```python
avg_returns_1yr = MF.groupby('category')['returns_1yr'].mean().sort_values()
plt.barh(avg_returns_1yr.index, avg_returns_1yr.values, color='lightgreen')
plt.xlabel('Average Returns (%)')
plt.ylabel('Category')
plt.title('Average Returns for Different Categories')
plt.show()
```

```python
plt.scatter(MF['fund_size_cr'], MF['expense_ratio'], alpha=0.5, color='orange')
plt.xlabel('Fund Size (in Crores)')
plt.ylabel('Expense Ratio')
plt.title('Fund Size vs. Expense Ratio')
plt.show()
```

```python
risk_level_counts = MF['risk_level'].value_counts()
plt.pie(risk_level_counts, labels=risk_level_counts.index, autopct='%1.2f%%', startangle=140)
plt.title('Distribution of Risk Levels')
plt.axis('equal')
plt.show()
```

```python
avg_ratings = MF.groupby('amc_name')['rating'].mean().sort_values()
plt.bar(avg_ratings.index, avg_ratings.values, color='lightblue')
plt.xlabel('AMC Name')
plt.ylabel('Average Rating')
plt.title('Average Ratings by AMC Name')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.plot(MF['fund_age_yr'], MF['returns_1yr'], marker='o', color='purple')
plt.xlabel('Fund Age (in Years)')
plt.ylabel('Returns over 1 Year (%)')
plt.title('Returns over 1 Year vs. Fund Age')
plt.grid(True)
plt.show()
```

```python
sns.boxplot(x=MF['category'], y=MF['rating'])
plt.xlabel('Category')
plt.ylabel('Rating')
plt.title('Ratings by Category')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.pairplot(MF[['fund_size_cr', 'expense_ratio', 'rating']])
plt.suptitle('Pairplot of Fund Size, Expense Ratio, and Ratings')
plt.show()
```

```python
avg_returns_5yr = MF.groupby('sub_category')['returns_5yr'].mean().sort_values()
plt.bar(avg_returns_5yr.index, avg_returns_5yr.values, color='salmon')
plt.xlabel('Sub-category')
plt.ylabel('Average Returns over 5 Years (%)')
plt.title('Average Returns over 5 Years by Sub-category')
plt.xticks(rotation=90)
plt.show()
```
